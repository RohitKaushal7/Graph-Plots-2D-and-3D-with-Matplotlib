```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import math
pi = math.pi
```

```python
# Message Signal
carrier_frequency = 100
message_frequency = 10
x = np.arange(0, pi, 0.0001)
y = np.cos(message_frequency*x)
plt.plot(x, y)
plt.title('Message Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi], ['0', '\u03C0/2', '\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9c*t)")
plt.grid(True, which = 'both')
plt.show()
```

```python
# Carrier Signal
carrier_frequency = 100
message_frequency = 10
x = np.arange(0, pi, 0.000001)
y = np.cos(carrier_frequency*x)
plt.plot(x, y)
plt.title('Message Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi], ['0', '\u03C0/2', '\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9m*t)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#Amplitude Modulation
carrier_frequency = 100
message_frequency = 10
x = np.arange(0, pi/2, 0.00001)
y = np.cos(message_frequency*x)*np.cos(carrier_frequency*x)
plt.plot(x, y)
plt.title('AM Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2], ['0', '\u03C0/2'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9m*t)*cos(\u03A9c*t)")
plt.grid(True, which = 'both')
plt.show()
```
